**Question No 1:**
`How many total pageview events did the links in the provided dataset receive in the full period? How many per day?`

```python
# Dispaly the Unique Value of event colm
eventColm = dataset['event'].unique()
print(f"Here is the Unique Value of Event Colm :{eventColm}")
```

```python
# Dispaly the Unique Value of date colm
dateColm = dataset['date'].unique()
print(f"Here is the Unique Value of Date Colm :{dateColm}")
```

```python
# Calculate the Sum of Total PageView
totalPageView = dataset["event"].value_counts()["pageview"]

# Find out the Number of Days
startDate = dataset["date"].min()
endDate   = dataset["date"].max()
days      = int(endDate.split("-")[2]) - int(startDate.split("-")[2]) + 1 #(Add the Starting and Ending Value also)

# Total PageView in the Per Day
pageviewPerDay = totalPageView / days

# Display the Answer
print(f"Here is the Total PageView According to All Date : {totalPageView}")
print(f"Here is the Total Days in the Dataset : {days}")
print(f"Here is the Total PageView in the Per Day : {pageviewPerDay}")
```

**Question No 2:**
`What about the other recorded events?`

```python
# Calculate the Sum of Click & PreView
totalClick   = dataset["event"].value_counts()["click"]
totalPreView = dataset["event"].value_counts()["preview"]

# Find out the Number of Days
startDate = dataset["date"].min()
endDate   = dataset["date"].max()
days      = int(endDate.split("-")[2]) - int(startDate.split("-")[2]) + 1 #(Add the Starting and Ending Value also)

# Total (Click & PreView) in the Per Day
clickPerDay   = totalClick / days
previewPerDay = totalPreView / days

# Display the Answer
print(f"Here is the Days in the Dataset : {days}")
print(f"Here is the Total Click According to all Date : {totalClick}")
print(f"Here is the Total Click in the Per Day : {clickPerDay}")
print(f"Here is the Total PreView According to all Date : {totalPreView}")
print(f"Here is the Total PreView in the Per Day : {previewPerDay}")
```

**Question No 4:**
`What was the overall click rate (clicks/pageviews)?`

```python
# Claculate the Overall Click Rate
totalClick   = dataset["event"].value_counts()["click"]
totalPreView = dataset["event"].value_counts()["preview"]

clickRate = totalClick / totalPreView

# Display the Answer
print(f"Here is the Overall Click Rate : {clickRate}")
```

**Question No 5:**
`How does the clickrate distribute across different links?`

```python
# Count the Click and PageView on per LinkId
click    = dataset[dataset["event"] == "click"].groupby("linkid").size().rename("Click")
pageview = dataset[dataset["event"] == "pageview"].groupby("linkid").size().rename("PageView")

# Concatenate the These Resulst
clickData = pd.concat([click , pageview] ,  axis = 1).fillna(0)

# Calculate the ClickRate for Each Link
clickData["clickRate"] = clickData["Click"] / clickData["PageView"]

# Display the Answer
clickData.head(20)
```
